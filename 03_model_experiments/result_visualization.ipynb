```python
import os
import pandas as pd
import matplotlib.pyplot as plt

def plot_training_history(backbone_name: str, loss_name: str, checkpoint_dir: str):
    """
    특정 백본(backbone) + 손실 함수(loss) 조합의 학습 히스토리를 시각화합니다.

    Args:
        backbone_name (str): 예) "EfficientNet-B3"
        loss_name (str): 예) "BatchHardTriplet" / "SemiHardTriplet" / "InfoNCE"
        checkpoint_dir (str): history CSV가 저장된 디렉토리 (기본값: CHECKPOINT_DIR)

    Example:
        plot_training_history("EfficientNet-B3", "BatchHardTriplet")
    """
    # CSV 경로 설정
    history_file = os.path.join(checkpoint_dir, f"{backbone_name}_{loss_name}_history.csv")

    if not os.path.exists(history_file):
        print(f"❌ 파일을 찾을 수 없습니다: {history_file}")
        return

    # 2데이터 로드
    df = pd.read_csv(history_file)
    print(df)

    # 시각화
    plt.figure(figsize=(10, 6))

    # Train Loss
    if "train_loss" in df.columns:
        plt.plot(df["epoch"], df["train_loss"], label="Train Loss", marker="o", color="tab:blue")

    # Validation Recall@K
    for k, color in zip(["val_R@1", "val_R@5", "val_R@10"], ["tab:orange", "tab:green", "tab:red"]):
        if k in df.columns:
            plt.plot(df["epoch"], df[k], label=k, marker="o", color=color)

    plt.xlabel("Epoch", fontsize=12)
    plt.ylabel("Value", fontsize=12)
    plt.title(f"{backbone_name} + {loss_name} Training History", fontsize=14, fontweight="bold")
    plt.grid(True, linestyle="--", alpha=0.6)
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
CHECKPOINT_DIR = os.path.join(MODULE_PATH, "checkpoints_1")

plot_training_history("EfficientNet-B3", "semiHardTriplet", CHECKPOINT_DIR)
```
